## import libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns 
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

## Load DataSet 

```python
df = pd.read_csv("F:/regression project/boston.csv")
df.head()
```

## preprocessing to ensure the quality of data

```python
# dataTypes of variables 
df.info()
```

```python
# Check for missing values 
df.isnull().sum()
```

```python
# Rename a MEDV to PRICE 
df = df.rename(columns={'MEDV': 'PRICE'})
```

```python
df.head()
```

## EDA

```python
corr = df.corr()
corr
```

```python
def plotting_scatter():
    features = [col for col in df.columns if col != 'PRICE']
    for feature in features:
        sns.regplot(x=feature,y='PRICE',data=df)
        plt.show()
plotting_scatter()
```

```python
def Heat_Map():
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
    plt.title('Correlation Matrix Heatmap')
    plt.tight_layout()
    plt.show()
Heat_Map()
```

```python
# Define a function to create a box plot for each column
def plot_outliers_boxplot(data):
    # Create a figure with subplots for each column
    fig, axs = plt.subplots(ncols=len(data.columns), figsize=(20, 5))

    # Create a box plot for each column
    for i, col in enumerate(data.columns):
        sns.boxplot(x=data[col], ax=axs[i])
        axs[i].set_xlabel(col)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_outliers_boxplot(df)
```

## Standarizing The Data 

```python
# spliting the data 
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
x.head()
```

```python
# train test split 
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=52)
```

```python
#standarize the data 
Scaler = StandardScaler()
x_train = Scaler.fit_transform(x_train)
x_test = Scaler.transform(x_test)
```

## build linear regression model 

```python
Linear_Regression = LinearRegression()
Linear_Regression.fit(x_train,y_train)
print(f"The is the coeffient of independent {Linear_Regression.coef_}\n")
print(f"The is the Y_intercept of independent {Linear_Regression.intercept_}")
```

```python
Regression_predict = Linear_Regression.predict(x_test)
```

```python
Regression_predict
```

```python
sns.regplot(x=Regression_predict,y=y_test)
plt.show()
```

```python
error = y_test - Regression_predict
sns.displot(error,kind='kde')
```

```python
## error fucations
print("the mean_squared_error is ",mean_squared_error(y_test,Regression_predict))
print("the r square is ",np.sqrt((mean_squared_error(y_test,Regression_predict))))
print("the mean_absolute_error is ",mean_absolute_error(y_test,Regression_predict))
```

```python
Score = r2_score(y_test,Regression_predict)
print(round(Score,4))
```
